```python
%load_ext autoreload
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tools import clean_data, get_pca_features, get_features
```

```python
pd.set_option('display.max_columns', None)
train_df = pd.read_pickle("train_test_hackathon.pkl")
train_df = clean_data(train_df)
```

```python
plt.hist(train_df['Feature_1_cv_enum'])
plt.title("Feature 1 distribution")
plt.show()
print(train_df['Feature_1_cv_enum'].value_counts())

plt.hist(train_df['Feature_23_adv_cv_enum'])
plt.title("Feature 23 distribution")
plt.show()
print(train_df['Feature_23_adv_cv_enum'].value_counts())
```

```python
fig, axs = plt.subplots(1, 3, figsize=(5, 5), sharey=True, sharex=True)
fig.set_figheight(5)
fig.set_figwidth(15)

# Target R
axs[0].scatter(train_df[train_df['target_r'] == 1]['Feature_2'], train_df[train_df['target_r'] == 1]['Feature_3'], color='blue')
axs[0].scatter(train_df[train_df['target_r'] == 0]['Feature_2'], train_df[train_df['target_r'] == 0]['Feature_3'], color='red')
axs[0].set_title("Target R")

# Target G
axs[1].scatter(train_df[train_df['target_g'] == 1]['Feature_2'], train_df[train_df['target_g'] == 1]['Feature_3'], color='blue')
axs[1].scatter(train_df[train_df['target_g'] == 0]['Feature_2'], train_df[train_df['target_g'] == 0]['Feature_3'], color='red')
axs[1].set_title("Target G")

# Target B
axs[2].scatter(train_df[train_df['target_b'] == 1]['Feature_2'], train_df[train_df['target_b'] == 1]['Feature_3'], color='blue')
axs[2].scatter(train_df[train_df['target_b'] == 0]['Feature_2'], train_df[train_df['target_b'] == 0]['Feature_3'], color='red')
axs[2].set_title("Target B")

# plt.ylim(0, 0.1 * 1e10)
# plt.xlim(0, 0.1 * 1e9)
plt.xlabel("Feature 2")
plt.ylabel("Feature 3")

plt.show()
```

```python
features_1 = train_df[['Feature_2', 'Feature_3', 'Feature_4', 'Feature_5', 'Feature_6', 'Feature_7', 'Feature_8', 'Feature_9', 'target_r']]
features_1.dropna(inplace=True)
sns.pairplot(features_1, hue="target_r")

features_2 = train_df[['Feature_2', 'Feature_3', 'Feature_4', 'Feature_5', 'Feature_6', 'Feature_7', 'Feature_8', 'Feature_9', 'target_g']]
features_2.dropna(inplace=True)
sns.pairplot(features_2, hue="target_g")

features_3 = train_df[['Feature_2', 'Feature_3', 'Feature_4', 'Feature_5', 'Feature_6', 'Feature_7', 'Feature_8', 'Feature_9', 'target_b']]
features_3.dropna(inplace=True)
sns.pairplot(features_3, hue="target_b")
```
